```python
import pandas as pd
import torch as t
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
data=pd.read_csv(r"C:\Users\hrish\Projects\Deep_learning_projects\Trigram\NationalNames.csv")
```

```python
names=data['Name'].unique()
```

```python
random.seed(42)
names=list(names)
random.shuffle(names)
```

```python
names=[n.lower() for n in names]
```

```python
alpha=list(set(''.join(names)))
alpha.sort()
```

```python
con={a:i+1 for i,a in enumerate(alpha)}
con['.']=0
```

```python
X=[]
y=[]
for nm in names:
    ct=[0,0,0]
    nm=nm+'.'
    for c in nm:
        X.append(ct)
        y.append(con[c])
        ct=ct[1:]+[con[c]]
```

### Building the NN

```python
g = t.Generator().manual_seed(2147483647)
C = t.randn((27, 10), generator=g)
W1 = t.randn((30, 500), generator=g)
b1 = t.randn(500, generator=g)
W2 = t.randn((500, 27), generator=g)
b2 = t.randn(27, generator=g)
parameters = [C, W1, b1, W2, b2]
```

```python
for p in parameters:
    p.requires_grad=True
```

```python
Xtr=t.tensor(X[:500001])
ytr=t.tensor(y[:500001])
Xts=t.tensor(X[500001:])
yts=t.tensor(y[500001:])
```

```python
Xtr.shape[0]
```

```python
emb=C[Xtr[:5]]
```

```python
emb.view(-1,C.shape[1]*Xtr.shape[1])
```

```python
lre = t.linspace(-3, 1, 1000)
lrs = 10**lre
```

```python
lossi=[]
lri=[]
lr=1
for i in range(40000):
    ix=t.randint(0,Xtr.shape[0],(1000,))
    emb=C[Xtr[ix]] #1000*3*10
    t1=emb.view(-1,C.shape[1]*Xtr.shape[1])@W1+b1 # 1000*30 * 30*500 
    t1=t.tanh(t1)
    logits=t1@W2+b2 # 1000*500 * 500*27
    loss=t.nn.functional.cross_entropy(logits,ytr[ix])
    for p in parameters:
        p.grad=None
    loss.backward()
    if(i>20000):
        lr=0.1
    for p in parameters:
        p.data-=lr*p.grad
    #lossi.append(loss.item())
    #lri.append(lre[i])
```

```python
rev={j:i for i,j in con.items()}
```

```python
for i in range(50):
    cnt=[0,0,0]
    out=[]
    while True:
        emb=C[t.tensor(cnt)]
        t1=emb.view(-1,C.shape[1]*Xtr.shape[1])@W1+b1 # 1000*30 * 30*500 
        t1=t.tanh(t1)
        logits=t1@W2+b2 # 1000*500 * 500*27
        probs = t.nn.functional.softmax(logits, dim=1)
        ix = t.multinomial(probs, num_samples=1, generator=g).item()
        out.append(rev[ix])
        cnt=cnt[1:]+[ix]
        if(ix==0):
            break
    print(''.join(out))
```

### Building Tetragon

```python
l=0
d={}
for i in range(27):
    for j in range(27):
        for k in range(27):
            d[((i,j),k)]=l
            l+=1
```

```python
wg=t.randn((19683,27),requires_grad=True)
```

```python
wx=t.tensor([d[((int(x[2]),int(x[1])),int(x[0]))] for x in Xtr])
```

```python
for i in range(1000):
    logits=wg[wx]
    loss=t.nn.functional.cross_entropy(logits,ytr)
    wg.grad=None
    loss.backward()
    wg.data-=50*wg.grad
```

```python
loss
```

```python
emb=C[Xtr]
t1=emb.view(-1,C.shape[1]*Xtr.shape[1])@W1+b1
t1=t.tanh(t1)
logits1=t1@W2+b2
probs1 = t.nn.functional.softmax(logits1, dim=1)
logits2=wg[t.tensor([d[((int(x[2]),int(x[1])),int(x[0]))] for x in Xtr])]
probs2= t.nn.functional.softmax(logits2, dim=1)
probs=(probs1+probs2)/2
loss=-t.log(probs[t.arange(len(ytr)),ytr]).mean()
```

```python
loss
```

```python
for i in range(50):
    out=[]
    ctx=[0,0,0]
    while True:
        emb=C[t.tensor(ctx)]
        t1=emb.view(-1,C.shape[1]*Xtr.shape[1])@W1+b1
        t1=t.tanh(t1)
        logits1=t1@W2+b2
        probs1 = t.nn.functional.softmax(logits1, dim=1)
        logits2=wg[t.tensor([d[((int(ctx[2]),int(ctx[1])),int(ctx[0]))]])]
        probs2= t.nn.functional.softmax(logits2, dim=1)
        probs=(probs1+probs2)/2
        ix=t.multinomial(probs, num_samples=1, generator=g).item()
        out.append(rev[ix])
        ctx=ctx[1:]+[ix]
        if ix==0:
            break
    print(''.join(out))
```
